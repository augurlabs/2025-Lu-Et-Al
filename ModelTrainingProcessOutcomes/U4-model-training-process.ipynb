# Read needed dataset

```python
U4_New_Combine_Data = pd.read_csv('U4NewCombinedFeatureswithInterFactors1.csv')
```

```python
list(U4_New_Combine_Data.columns)
```

```python
# Machine learning systems work with integers, we need to encode these
# string characters into ints

encoder = LabelEncoder()

# Now apply the transformation to all the columns:
for col in U4_New_Combine_Data.columns:
    U4_New_Combine_Data[col] = encoder.fit_transform(U4_New_Combine_Data[col])

y_label = U4_New_Combine_Data['POSTU4']
X_features = U4_New_Combine_Data.drop(['playerName','POSTU4'],axis=1)
```

```python
y_label
```

```python
X_features
```

```python
# Standardized the features
X_features_stdPre = StandardScaler().fit_transform(X_features)

# Feature selection
model = ExtraTreesClassifier()
random.seed(0)
model.fit(X_features_stdPre, y_label)
print(model.feature_importances_)
# Calculate the importance rate for feature selection
feat_importances = pd.Series(model.feature_importances_, index=X_features.columns)
```

```python
# Involve needed resampling method
resampling_method = {
    'RandomOver': ros,
    'Smote': smote,
    'adasym': adasym,
    'borderline': borderline,
    'SmoteN': sampler,
    'SmoteENN': smote_enn,
    'SmoteTomek': smote_tomek,
    'clusterCen': cc,
    'RandomUnder': rus,
    'RandomUnderReSam': rus1,
    'NearMiss': nm1,
    'NearestNeighbor': enn,
    'NearestNeighborAll': enn1,
    'NearestNeighborMode': enn2,
    'RepeatedNearNeighbor': renn,
    'allKNN': allknn,
    'CondensedNearNeighbor': cnn,
    'OnesidedSelection': oss,
    'NeighborCleaningRule': ncr,
    'InstanceHardness': iht
}
```

```python
# Involve machine learning algorithms for the ensemble
model_fit_algorithm = {
    'SVC': svc_clf,
    'LinearSVC': linsvc_clf,
    'RFM': rf_clf,
    'GaussianNB': gnb_clf,
    'DTM': dt_clf,
    'XGBoost': xg_clf,
    'LogiReg': logreg_clf,
    'KNN': knn,
    'Nsvc': Nsvc_clf,
    'decisionTree': dt_clf,
    'AdaBoost': AdaBoost_clf,
    'gradientB': gradientB_clf,
    'linearD': linearD_clf,
    'quadraticDA': quadraticDA_clf,
    'baggingC': baggingC_clf,
    'extraTree': extraTree_clf,
    'gradientBoosting': gradientB_clf
}
```

```python
ensemble_model = VotingClassifier(estimators=list(model_fit_algorithm.items()), voting='hard')
```

```python
if not os.path.exists('U4ModelTrainingResults.csv'):
    outcome = pd.DataFrame(columns=['Feature Number', 'Iteration Number', 'Resampling Name', 
                                    'Test Accuracy', 'Macro Precision', 'Macro Recall', 'Macro F1', 
                                    'High Precision', 'High Recall', 'High F1', 
                                    'Low Precision', 'Low Recall', 'Low F1'])
    outcome.to_csv('U4ModelTrainingResults.csv', index=False)
```

```python
# Metrics lists
featureNumber = []
iterationNumber = []
resampleMethod = []
modelName = []
testAccuracy = []
macroPrecision = []
macroRecall = []
macroF1 = []
highPrecision = []
highRecall = []
highF1 = []
lowPrecision = []
lowRecall = []
lowF1 = []

# Start the model training
for x in range(2, 177):
    LargestVar = pd.DataFrame(feat_importances.nlargest(x))
    cols = LargestVar.index
    X_features_1 = X_features[cols]

    # Prepare the dataset and parameters
    X_features_std = StandardScaler().fit_transform(X_features_1)
    # Parameters
    dim = len(X_features_std[0])
    n_classes = len(np.unique(y_label))
    
    for z in range(1, 16):
        random_state = z
        # Split into train/test
        X_train, X_test, y_train, y_test = train_test_split(X_features_std, y_label, test_size=0.2, stratify=y_label, random_state=random_state)

        
        try:
            for i, (Rname, method) in enumerate(resampling_method.items()):
                X_resample, Y_resample = method.fit_resample(X_train, y_train)
                # Fit the ensemble model
                ensemble_model.fit(X_resample, Y_resample)
                pred = ensemble_model.predict(X_test)
                class_report = classification_report(y_test, pred, target_names=['High', 'Low'], output_dict=True)
                
                # Extract metrics
                macro_precision = class_report['macro avg']['precision']
                macro_recall = class_report['macro avg']['recall']
                macro_f1 = class_report['macro avg']['f1-score']
                High_precision = class_report['High']['precision']
                High_recall = class_report['High']['recall']
                High_f1 = class_report['High']['f1-score']
                Low_precision = class_report['Low']['precision']
                Low_recall = class_report['Low']['recall']
                Low_f1 = class_report['Low']['f1-score']

                model_acc = accuracy_score(y_test, pred)
                print(model_acc)

                # Append metrics to lists
                featureNumber.append(x)
                iterationNumber.append(z)
                resampleMethod.append(Rname)
                testAccuracy.append(model_acc)
                macroPrecision.append(macro_precision)
                macroRecall.append(macro_recall)
                macroF1.append(macro_f1)
                highPrecision.append(High_precision)
                highRecall.append(High_recall)
                highF1.append(High_f1)
                lowPrecision.append(Low_precision)
                lowRecall.append(Low_recall)
                lowF1.append(Low_f1)
                outcome = pd.DataFrame([[x, z, Rname, model_acc, macro_precision, macro_recall, macro_f1, 
                                        High_precision, High_recall, High_f1, Low_precision, Low_recall, Low_f1]],
                                        columns=['Feature Number', 'Iteration Number', 'Resampling Name', 
                                                 'Test Accuracy', 'Macro Precision', 'Macro Recall', 'Macro F1', 
                                                 'High Precision', 'High Recall', 'High F1', 
                                                 'Low Precision', 'Low Recall', 'Low F1'])
                
                # Append results to CSV file
                with open('U4ModelTrainingResults.csv', 'a') as f:
                    outcome.to_csv(f, header=False, index=False)

        except Exception:
            print("An error occurred, moving on to the next iteration")
```
